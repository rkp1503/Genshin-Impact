```python
import json

import utils.sub_stat_helper as ssh
import utils.sub_stat_single as sss
import utils.sub_stat_double as ssd
import utils.sub_stat_triple as sst
import utils.sub_stat_quad as ssq

################################################################################

with open("assets/data/sub_stats.json", "r") as file:
    json_data: dict = json.load(file)
    pass

all_sub_stats: list[str] = [stat for stat in json_data.keys()]

normalize = True
```

## Single Sub Stat Distribution

```python
sub_stat_1: str = all_sub_stats[8]
sub_stats: list[str] = [sub_stat_1]

data: list[list[float]] = sss.get_all_possible_rolls(json_data, sub_stats)
roll_values: list[float] = ssh.convert_data_to_roll_value(json_data, sub_stats, data, normalize=normalize)
ssh.display_distribution(sub_stats, roll_values, normalize=normalize)
```

## Double Sub Stat Distribution

```python
sub_stat_2: str = all_sub_stats[9]
sub_stats: list[str] = [sub_stat_1, sub_stat_2]

data: list[list[float]] = ssd.get_all_possible_rolls(json_data, sub_stats)
roll_values: list[float] = ssh.convert_data_to_roll_value(json_data, sub_stats, data, normalize=normalize)
ssh.display_distribution(sub_stats, roll_values, normalize=normalize, bw_adjust=2)
```

## Triple Sub Stat Distribution

```python
sub_stat_3: str = all_sub_stats[7]
sub_stats: list[str] = [sub_stat_1, sub_stat_2, sub_stat_3]

data: list[list[float]] = sst.get_all_possible_rolls(json_data, sub_stats)
roll_values: list[float] = ssh.convert_data_to_roll_value(json_data, sub_stats, data, normalize=normalize)
ssh.display_distribution(sub_stats, roll_values, normalize=normalize, bw_adjust=3)
```
